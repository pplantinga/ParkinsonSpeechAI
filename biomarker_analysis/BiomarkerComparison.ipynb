```python
import numpy as np

def cohend(d1, d2) -> float:
    d1, d2 = np.array(d1), np.array(d2)

    # calculate the size of samples
    n1, n2 = len(d1), len(d2)

    # calculate the variance of the samples
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)

    # calculate the pooled standard deviation
    s = np.sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))

    # calculate the means of the samples
    u1, u2 = np.mean(d1), np.mean(d2)

    # return the effect size
    return (u1 - u2) / s

def print_cohens_per_metric(df):
    for dataset in df["dataset"].unique():
        print(dataset)
        print("-" * len(dataset))
        data = df.filter(df["dataset"].eq(dataset))
        pdata = data.filter(data["status"].eq("PD"))
        cdata = data.filter(data["status"].eq("HC"))
        for metric in ["VOT", "DPI", "F0SD", "SFSB", "UPR", "GNE"]:
            print(f"{metric} cohen d={cohend(cdata[metric], pdata[metric]):.3f}")
        print("")


print_cohens_per_metric(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pdata = read_df.filter(read_df["status"].eq("PD"))
cdata = read_df.filter(read_df["status"].eq("HC"))
fig = plt.figure(figsize=(6, 3))
axes = fig.subplots(nrows=2, ncols=2, sharey=False, sharex=False, gridspec_kw = {'wspace':0.05, 'hspace':0.5}).ravel()
for i, metric in enumerate(["VOT", "DPI", "F0SD", "SFSB"]):
    effect_size = cohend(cdata[metric], pdata[metric])
    #print(metric, "cohen d=", effect_size)
    sns.kdeplot(data=read_df, x=metric, hue="status", clip=(0, None), ax=axes[i], hue_order=["HC", "PD"], lw=3)
    axes[i].legend(["Patients", "Controls"])
    #axes[i].set_title(metric.title())
    axes[i].locator_params(nbins=5, axis='x')
    axes[i].tick_params(direction='in')
    axes[i].set_yticklabels([])
    axes[i].set_ylabel(None)
    ymin, ymax = axes[i].get_ylim()
    xmin, xmax = axes[i].get_xlim()
    axes[i].text(x=xmax - xmax / 10, y=ymax / 5, horizontalalignment="right", s=f"d= {effect_size:.2f}")
#fig.suptitle("Reading Task: 4 Interpretable Scores")
fig.savefig("reading-task-scores-kernel.pdf", bbox_inches="tight")
```

```python
metrics = ["GNE", "VOT", "DPI", "F0SD", "SFSB"]
metric_labels = ["GNE", "VOT", "DPI", "F$_0$SD", "SFSB"]
def collect_cohend_metric(data):
    pdata = data.filter(data["status"].eq("PD"))
    cdata = data.filter(data["status"].eq("HC"))
    results = {}
    for metric in metrics:
        sign = -1 if metric in ["GNE", "DPI"] else 1
        results[metric] = sign * cohend(cdata[metric], pdata[metric])
    return results

rows = {
    'QPN-L1': collect_cohend_metric(df.filter(df["dataset"].eq("qpn") & df["first_lang"] == True)),
    'QPN-L2+': collect_cohend_metric(df.filter(df["dataset"].eq("qpn") & df["first_lang"] == False)),
    'MVDR-KCL': collect_cohend_metric(df.filter(df["dataset"].eq("mvdr_kcl"))),
    'PC-GITA': collect_cohend_metric(df.filter(df["dataset"].eq("pc_gita"))),
    'Spontaneous': collect_cohend_metric(spont_df),
    'Constrained': collect_cohend_metric(read_df),
}

# Plotting
plt.figure(figsize=(3, 2))
# Use 'RdBu_r' or 'viridis' for Nature-style colors
ax = sns.heatmap(pl.DataFrame(list(rows.values())), xticklabels=metric_labels, yticklabels=list(rows.keys()), annot=True, cmap='RdBu', center=0, 
                 cbar_kws={'label': "Cohen's $d$ Effect Size"}, fmt=".2f", linewidths=.5, annot_kws={"weight": "bold"})

#plt.title('A. Biomarker Effect Size across Datasets and Tasks', loc='left', fontweight='bold', pad=20)
#plt.ylabel('Condition / Dataset')
#plt.xlabel('Biomarker')
#plt.tight_layout()
plt.savefig('biomarker_effect_size.pdf', bbox_inches="tight")
```

```python
def plot_task_biomarker_grid(df):
    # 1. Categorize Tasks
    # Mapping based on your filter logic
    constrained_tasks = ["read", "ReadText", "read text", "repeat"]
    spontaneous_tasks = ["dpt", "SpontaneousDialogue", "monologue", "recall"]
    
    def categorize_task(task):
        if task in constrained_tasks: return "Constrained"
        if task in spontaneous_tasks: return "Spontaneous"
        return None

    df['task_type'] = df['task'].apply(categorize_task)
    # Drop rows that don't fall into either category if necessary
    plot_df = df.dropna(subset=['task_type'])

    # 2. Setup Figure
    # Rows: Measures (F0SD, SFSB) | Columns: Task Types (Spont, Constrained)
    fig, axes = plt.subplots(2, 2, figsize=(4,2.5), sharex=False)
    
    palette = {"PD": "#E66414", "HC": "#00A8A8"}
    measures = ['F0SD', 'SFSB']
    task_types = ['Spontaneous', 'Constrained']

    for j, measure in enumerate(measures):
        for i, task_type in enumerate(task_types):
            ax = axes[i, j]
            subset = plot_df[plot_df['task_type'] == task_type]
            
            # Draw KDE
            sns.kdeplot(data=subset, x=measure, hue='status', palette=palette,
                        fill=True, ax=ax, alpha=0.3, common_norm=False, clip=(0, None))
            
            # 3. Calculate Cohen's d for the title
            pd_vals = subset[subset['status'] == 'PD'][measure].dropna()
            ctrl_vals = subset[subset['status'] == 'HC'][measure].dropna()
            
            if len(pd_vals) > 5 and len(ctrl_vals) > 5:
                d = cohend(ctrl_vals, pd_vals)
                d_label = f" ($d={d:.2f}$)"
            else:
                d_label = ""

            # Formatting
            #ax.set_title(f"{measure} | {task_type}{d_label}", loc='left', fontweight='bold')
            #ax.set_ylabel('Density')
            sns.despine(ax=ax)
            #ax.grid(True, linestyle='--', alpha=0.2)
            ax.set_yticks([])
            
            if measure == "SFSB":
                ax.set_xlim([0.02, 0.12])
                ax.set_xlabel(f"SFSB | {task_type}{d_label}")
            else:
                ax.set_xlim([0, 80])
                ax.set_xlabel(f"F$_0$SD | {task_type}{d_label}")

            

    # Clean up layout
    plt.tight_layout()
    plt.savefig('biomarker_task_comparison.pdf', bbox_inches='tight')
    plt.show()

plot_task_biomarker_grid(df.to_pandas())
```
