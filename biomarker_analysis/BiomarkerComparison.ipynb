```python
read_df = df.filter(df["task"].is_in(["read", "ReadText", "read text", "repeat"]))
```

```python
spont_df = df.filter(df["task"].is_in(["dpt", "SpontaneousDialogue", "monologue", "recall"]))
```

```python
import numpy as np

def cohend(d1, d2) -> float:
    d1, d2 = np.array(d1), np.array(d2)

    # calculate the size of samples
    n1, n2 = len(d1), len(d2)

    # calculate the variance of the samples
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)

    # calculate the pooled standard deviation
    s = np.sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))

    # calculate the means of the samples
    u1, u2 = np.mean(d1), np.mean(d2)

    # return the effect size
    return (u1 - u2) / s

def print_cohens_per_metric(df):
    for dataset in df["dataset"].unique():
        print(dataset)
        print("-" * len(dataset))
        data = df.filter(df["dataset"].eq(dataset))
        pdata = data.filter(data["status"].eq("PD"))
        cdata = data.filter(data["status"].eq("HC"))
        for metric in ["VOT", "DPI", "F0SD", "SFB", "UPR", "GNE"]:
            print(f"{metric} cohen d={cohend(cdata[metric], pdata[metric]):.3f}")
        print("")


print_cohens_per_metric(df)
```

```python
print_cohens_per_metric(read_df)
```

```python
print_cohens_per_metric(spont_df)
```

```python
def print_cohens_per_metric_overall(df):
    pdata = df.filter(df["status"].eq("PD"))
    cdata = df.filter(df["status"].eq("HC"))
    for metric in ["VOT", "DPI", "F0SD", "SFB", "UPR", "GNE"]:
        print(f"{metric} cohen d={cohend(cdata[metric], pdata[metric]):.3f}")
    print("")

print_cohens_per_metric_overall(df)
```

```python
print_cohens_per_metric_overall(read_df)
```

```python
print_cohens_per_metric_overall(spont_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pdata = read_df.filter(read_df["status"].eq("PD"))
cdata = read_df.filter(read_df["status"].eq("HC"))
fig = plt.figure(figsize=(10, 3))
axes = fig.subplots(nrows=2, ncols=2, sharey=False, sharex=False, gridspec_kw = {'wspace':0.05, 'hspace':0.5}).ravel()
for i, metric in enumerate(["VOT", "DPI", "F0SD", "SFB"]):
    effect_size = cohend(cdata[metric], pdata[metric])
    #print(metric, "cohen d=", effect_size)
    sns.kdeplot(data=read_df, x=metric, hue="status", clip=(0, None), ax=axes[i], hue_order=["HC", "PD"], lw=3)
    axes[i].legend(["Patients", "Controls"])
    #axes[i].set_title(metric.title())
    axes[i].locator_params(nbins=5, axis='x')
    axes[i].tick_params(direction='in')
    axes[i].set_yticklabels([])
    axes[i].set_ylabel(None)
    ymin, ymax = axes[i].get_ylim()
    xmin, xmax = axes[i].get_xlim()
    axes[i].text(x=xmax - xmax / 10, y=ymax / 5, horizontalalignment="right", s=f"d= {effect_size:.2f}")
fig.suptitle("Reading Task: 4 Interpretable Scores")
fig.savefig("reading-task-scores-kernel.pdf", bbox_inches="tight")
```

```python
pdata = spont_df.filter(spont_df["status"].eq("PD"))
cdata = spont_df.filter(spont_df["status"].eq("HC"))
fig = plt.figure(figsize=(10, 3.5))
axes = fig.subplots(nrows=2, ncols=2, sharey=False, sharex=False, gridspec_kw = {'wspace':0.05, 'hspace':0.6}).ravel()
xlabels = ["Voice Onset Delay (VOT)", "Duration of Pause Intervals", "F\u2080 Standard Deviation (F\u2080SD)", "Spectral Flux at Boundaries (SFB)"]
for i, metric in enumerate(["VOT", "DPI", "F0SD", "SFB"]):
    effect_size = cohend(cdata[metric], pdata[metric])
    print(metric, "cohen d=", effect_size)
    sns.kdeplot(data=spont_df, x=metric, hue="status", clip=(0, None), ax=axes[i], hue_order=["HC", "PD"], lw=3)
    axes[i].legend(["Patients", "Controls"])
    #axes[i].set_title(metric.title())
    axes[i].locator_params(nbins=5, axis='x')
    axes[i].tick_params(direction='in')
    axes[i].set_yticklabels([])
    axes[i].set_ylabel(None)
    axes[i].set_xlabel(xlabels[i])
    ymin, ymax = axes[i].get_ylim()
    xmin, xmax = axes[i].get_xlim()
    axes[i].text(x=xmax - xmax / 15, y=ymax / 5, ha="right", s=f"d= {effect_size:.2f}")
fig.suptitle("Spontaneous Speech Task: 4 Interpretable Scores")
fig.savefig("spontaneous-task-scores-kernel.pdf", bbox_inches="tight")
```

```python
import pingouin as pg

# Convert to pandas for pingouin (it doesn't support polars directly)
qpn_pandas = qpn.filter(qpn["task"] != "hbd").to_pandas()

# Calculate ICC
icc_result = pg.intraclass_corr(
    data=qpn_pandas, 
    targets='pid',    # subjects
    raters='task',    # tasks
    ratings='SFB'
)

print(icc_result)
```
